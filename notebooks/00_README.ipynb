## ğŸ“Š Dataset Background

**Adult Income Dataset** (UCI Machine Learning Repository)
- **Source**: 1994 US Census data
- **Scale**: 32,561 records, 15 features
- **Task**: Binary classification - predicting whether annual income exceeds $50K
- **Feature Types**:
  - Numerical features (6): age, education.num, hours.per.week, capital.gain/loss, fnlwgt
  - Categorical features (9): workclass, education, marital.status, occupation, relationship, race, sex, native.country, income

### âš ï¸ Special Value of This Dataset

This dataset is **not "clean data"** - it was deliberately chosen because:
1. **Has obvious data quality issues**: Missing values marked as "?", outliers, duplicates
2. **Has social biases**: 67% male gender distribution, 85% white race distribution
3. **Has class imbalance**: High income only accounts for 24%
4. **Has Simpson's Paradox risk**: Overall trends vs. stratified trends may be opposite

These "flaws" perfectly demonstrate **the necessity and value of data governance**!

### ğŸš¨ Biases in the Dataset

#### 1. Historical Bias
- **Gender imbalance**: 67% Male vs 33% Female
- **Race imbalance**: 85% White, underrepresentation of minorities
- **Root cause**: The 1994 social structure and job market itself was unequal

#### 2. Label Bias
- **Issue**: Income â‰  success, income â‰  ability
- **$50K threshold**: Median in 1994, but huge cost of living differences across regions
- **Ignores**: Non-monetary values (social contribution, job satisfaction)

#### 3. Representation Bias
- **Geographic**: Most samples from United States
- **Occupation**: Certain occupations severely missing (e.g., emerging tech roles)
- **Consequence**: Model reliability on minority groups is questionable
